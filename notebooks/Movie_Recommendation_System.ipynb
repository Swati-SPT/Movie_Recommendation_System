```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i["name"])
    return L
```

```python
movies1['genres']=movies1['genres'].apply(convert)
```

```python
movies1['genres'][0]
```

```python
movies1['keywords']=movies1['keywords'].apply(convert)
```

```python
def fetch_Cast(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i["name"])
    return L        
```

```python
movies1["cast"]=movies1["cast"].apply(fetch_Cast)
```

```python
movies1.head(2)
```

```python
def fetch_Dname(text):
    L=[]
    for i in ast.literal_eval(text):
        if i["job"]=="Director":
            L.append(i["name"])
    return L
```

```python
movies1["crew"]=movies1["crew"].apply(fetch_Dname)
```

```python
movies1.head(2)
```

```python
#we dont want space between words
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
    
```

```python
movies1["genres"]=movies1["genres"].apply(collapse)
movies1["crew"]=movies1["crew"].apply(collapse)
movies1["cast"]=movies1["cast"].apply(collapse)
movies1["keywords"]=movies1["keywords"].apply(collapse)
```

## Bag of words feature-extraction

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(lowercase=True,max_features=5000,stop_words='english',tokenizer=None)
```

```python
vector=cv.fit_transform(movies1['context']).toarray()
```

```python
vector.shape
```

```python
vector
```

## building recommendation system based on similarity matrix

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
movies1[movies1['title']=='Avatar'].index[0]
```

```python
def recommendation(movies):
    index=movies1[movies1['title']==movies].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(movies1.iloc[i[0]].title)
```

```python
recommendation('Avatar')
```

```python
recommendation('John Carter')
```
